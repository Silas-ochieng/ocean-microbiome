**Select a subset of columns for analysis**

```python
subset = df[['SampleID', 'Latitude', 'Longitude', 'SamplingDepth', 'Planctomycetacia', 'SAR202 clade']]
```

**Scatter Plot**

```python
import altair as alt

# Scatter plot
scatter = alt.Chart(df).mark_circle(size=60).encode(
    x='Longitude',
    y='Latitude',
    color='SamplingDepth',
    tooltip=['SampleID', 'Latitude', 'Longitude', 'SamplingDepth']
).properties(
    title='Ocean Microbiome Sampling Locations',
    width=600,
    height=400
)

scatter
```

**Multiple Coordinated Views**

**Geographic Distribution (Map View)**

```python
map_view = alt.Chart(df).mark_circle(size=60).encode(
    longitude='Longitude:Q',
    latitude='Latitude:Q',
    color='MarinePelagicBiomes:N',
    tooltip=['SampleID', 'Latitude', 'Longitude', 'MarinePelagicBiomes']
).properties(
    title='Marine Biomes Distribution',
    width=800,
    height=400
)

map_view
```

**Depth vs. Microbiome Abundance (Bar Chart)**

```python
bar_chart = alt.Chart(df).mark_bar().encode(
    x='SamplingDepth:Q',
    y='mean(Planctomycetacia):Q',
    color='MarinePelagicBiomes:N',
    tooltip=['SamplingDepth', 'mean(Planctomycetacia)']
).properties(
    title='Average Planctomycetacia Abundance by Sampling Depth',
    width=600,
    height=400
)

bar_chart
```

**Combine Views**

```python
# Define a selection
selection = alt.selection_point(fields=['MarinePelagicBiomes'], bind='legend')

# Apply the selection to the map
map_view = map_view.add_params(selection).encode(
    color=alt.condition(selection, 'MarinePelagicBiomes:N', alt.value('lightgray'))
)

# Filter bar chart based on the selection
bar_chart = bar_chart.transform_filter(selection)

# Combine the visualizations
combined = map_view & bar_chart
combined
```

**combine and save as an interactive html**

```python
combined.save('ocean_microbiome_visualization.html')

# Download the file to your local system
from google.colab import files
files.download('ocean_microbiome_visualization.html')
```
